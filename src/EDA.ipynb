```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("application_data.csv")
dataprev = pd.read_csv("previous_application.csv")
```

```python
print(f'La dimenci贸n del conjunto de datos application_data es {data.shape}')
print(f'La dimenci贸n del conjunto de datos previous_application es {dataprev.shape}')
```

```python
null_data = pd.DataFrame((data.isnull().sum())*100/data.shape[0]).reset_index()
null_data.columns = ['Columnas' , 'Porcentaje_de_valores_nulos']
```

```python
null_data_40 = null_data[null_data['Porcentaje_de_valores_nulos'] > 40]
null_data_40
```

```python
fig = plt.figure(figsize=(18,6))
ax = sns.pointplot(x="Columnas" , y="Porcentaje_de_valores_nulos" , data = null_data , color='green')
plt.xticks(rotation=90,fontsize=7)
ax.axhline(40, ls='--',color='red')
plt.title("Porcentaje de valores nulos en application data")
plt.ylabel("Porcentaje de valores nulos")
plt.xlabel("Columnas")
plt.show()
```

```python
null_data_prev = pd.DataFrame((dataprev.isnull().sum())*100/dataprev.shape[0]).reset_index()
null_data_prev.columns = ['Columnas' , 'Porcentaje_de_valores_nulos']
```

```python
null_dataprev_40 = null_data_prev[null_data_prev['Porcentaje_de_valores_nulos'] > 40]
null_dataprev_40
```

```python
fig = plt.figure(figsize=(18,6))
ax = sns.pointplot(x="Columnas" , y="Porcentaje_de_valores_nulos" , data = null_data_prev , color='orange')
plt.xticks(rotation=90,fontsize=7)
ax.axhline(40, ls='--',color='red')
plt.title("Porcentaje de valores nulos en previous application")
plt.ylabel("Porcentaje_de_valores_nulos")
plt.xlabel("Columnas")
plt.show()
```
